```python
d_ri = -0.3
d_r = 0
```

```python
def forced_photometry(img_r, img_i, 
        xi=coords_julen["xi"], eta=coords_julen["eta"], 
        update_coords=True, 
        ra0=ra0, dec0=dec0, 
        r_ap=1.7 * 2, 
        r_flux_max=10, auto=False, r_win=3, 
        backsize=64, backfiltsize=3, subtract_background=True,
        stack=stack,
        d_xi = d_xi, d_eta = d_eta,
        d_ri = d_ri, d_r = d_r,
        ):

    imgdata_r = img_r.data.astype(img_r.dtype.newbyteorder("="))
    imgdata_i = img_i.data.astype(img_i.dtype.newbyteorder("="))
    
    bkg_r = sep.Background(imgdata_r, bw=backsize, bh=backsize, fw=backfiltsize, fh=backfiltsize)
    bkg_i = sep.Background(imgdata_i, bw=backsize, bh=backsize, fw=backfiltsize, fh=backfiltsize)

    if subtract_background:
        img_nobkg_r = imgdata_r - bkg_r
        img_nobkg_i = imgdata_i - bkg_i
    else:
        img_nobkg_r = imgdata_r
        img_nobkg_i = imgdata_i

    
    coords_r = img_r.wcs.world_to_pixel(SkyCoord(ra0 + (xi+d_xi) / 60, dec0 + (eta+d_eta)/60, unit=u.deg))
    coords_i = img_i.wcs.world_to_pixel(SkyCoord(ra0 + (xi+d_xi) / 60, dec0 + (eta+d_eta)/60, unit=u.deg))

    if update_coords:
        coords_new_r = sep.winpos(img_nobkg_r, coords_r[0], coords_r[1], r_win)[0:2]
        coords_new_i = sep.winpos(img_nobkg_i, coords_i[0], coords_i[1], r_win)[0:2]

        for i in range(len(coords_new_r[0])):
            if np.sqrt((coords_new_r[0][i]-coords_r[0][i])**2 + (coords_new_r[1][i] - coords_r[1][i])**2) > r_flux_max:
                coords_new_r[0][i] = coords_r[0][i]
                coords_new_r[1][i] = coords_r[1][i]

            if np.sqrt((coords_new_i[0][i]-coords_i[0][i])**2 + (coords_new_i[1][i] - coords_i[1][i])**2) > r_flux_max:
                coords_new_i[0][i] = coords_i[0][i]
                coords_new_i[1][i] = coords_i[1][i]

    else:
        coords_new_r = coords_r
        coords_new_i = coords_i

    rh_r = sep.flux_radius(img_nobkg_r, coords_new_r[0], coords_new_r[1], np.full(len(xi), r_flux_max), 0.5)[0]
    rh_i = sep.flux_radius(img_nobkg_i, coords_new_i[0], coords_new_i[1], np.full(len(xi), r_flux_max), 0.5)[0]

    if auto:
        r = r_ap * rh_r
    else:
        r = r_ap
    flux_r, fluxerr_r, flag_r = sep.sum_circle(img_nobkg_r, coords_new_r[0], coords_new_i[1], r, err=bkg_r.globalrms, gain=1)

    if auto:
        r = r_ap * rh_i
    else:
        r = r_ap
    flux_i, fluxerr_i, flag_i = sep.sum_circle(img_nobkg_i, coords_new_i[0], coords_new_i[1], r, err=bkg_i.globalrms, gain=1)


    df = pd.DataFrame()

    zero_point_estimation_g = 27.93537 + stack * 2.5 * np.log10(3)
    zero_point_estimation_error_g = 0.01880
    
    zero_point_estimation_r = 28.06096 + stack * 2.5 * np.log10(3)
    zero_point_estimation_error_r = 0.01128
    
    zero_point_estimation_i = 27.49029 + stack * 2.5 * np.log10(3)
    zero_point_estimation_error_i = 0.01095


    airmass_r = img_r.header.get("AIRMASS", 0.0)
    airmass_i = img_i.header.get("AIRMASS", 0.0)
    extinction_coefficient_r = 0.07 
    extinction_coefficient_i = 0.04 


    df["x_r"] = coords_r[0]
    df["y_r"] = coords_r[1]
    df["x_i"] = coords_i[0]
    df["y_i"] = coords_i[1]
    df["rh_r"] = rh_r
    df["rh_i"] = rh_i

    df["x_r_win"] = coords_new_r[0]
    df["y_r_win"] = coords_new_r[1]
    df["x_i_win"] = coords_new_i[0]
    df["y_i_win"] = coords_new_i[1]
    df["flux_r"] =  flux_r
    df["flux_i"] = flux_i
    df["fluxerr_r"] =  fluxerr_r
    df["fluxerr_i"] = fluxerr_i

    
    df["mag_inst_r"] = -2.5 * np.log10(flux_r)
    df["mag_inst_i"] = -2.5 * np.log10(flux_i)
    
    df["mag_r"] = zero_point_estimation_r + df["mag_inst_r"] - (extinction_coefficient_r * airmass_r) + d_r
    df["mag_i"] = zero_point_estimation_i + df["mag_inst_i"] - (extinction_coefficient_i * airmass_i) - d_ri + d_r

    df["mag_ri"] = df["mag_r"] - df["mag_i"] + d_ri
    df['mag_r_ps'] = (df['mag_r'] - 0.011 - 0.125*(df['mag_ri']) 
                                           - 0.015 * (df['mag_ri']** 2))

    df['mag_i_ps'] = (df['mag_i'] + 0.001 - 0.006*(df['mag_ri'])
                                           - 0.002 * (df['mag_ri'] ** 2))

    # df['mag_i_ps'] = (df['mag_i'] + 0.004 - 0.014*(df['mag_gr'])
    #                                        - 0.001 * (df['mag_gr'] ** 2))



    return df
```

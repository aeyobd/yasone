```python
import matplotlib.pyplot as plt
```

```python
from astropy.table import Table
from astropy.coordinates import SkyCoord
from astropy import units as u
```

```python
import numpy as np
import arya
```

```python
import sys
sys.path.append("../../analysis")
import ana_utils
```

```python
def load_catalogue(objname, catname):
    path = f"../{objname}/coadd_median_r/tests/{catname}.fits"
    tab = Table.read(path, format="fits", hdu=2)

    obs_props = ana_utils.get_obs_props(objname)
    coord0 = SkyCoord(obs_props["ra"], obs_props["dec"], unit=u.degree)

    coords = SkyCoord(tab["ALPHA_J2000"], tab["DELTA_J2000"], unit=u.degree)
    tab["xi"], tab["eta"] = ana_utils.to_tangent(coords, coord0)
    
    return tab
```

```python
def plot_catalogue(cat, **kwargs):
    plt.xlabel(r"$\xi$ / arcmin")
    plt.ylabel(r"$\eta$ / arcmin")
    plt.scatter(60*cat["xi"], 60*cat["eta"], s=1, lw=0, **kwargs)
    plt.gca().invert_xaxis()
    plt.gca().set_aspect(1)
    print(len(cat))
```

```python
objname = "yasone3"
```

```python
for catname in ["min_area_1", "defaults", "min_area_6", "min_area_12"]:
    cat = load_catalogue(objname, catname)
    plot_catalogue(cat, label=catname)
    arya.Legend(-1)
```

```python
for catname in ["1sigma", "2sigma", "3sigma", "5sigma"]:
    cat = load_catalogue(objname, catname)
    plot_catalogue(cat)
    plt.title(catname)
    plt.show()
```

```python
def get_fwhm(catname):
    cat = Table.read(catname, format="fits", hdu=2)
    good_filt = cat["FLAGS"] == 0
    good_filt &= cat["ELLIPTICITY"] < 0.3
    good_filt &= cat["THRESHOLD"] > 50
    cat_good = cat[good_filt]

    fwhm = sigma_clipped_stats(cat_good["FWHM_WORLD"].to("arcsec"), stdfunc="mad_std")[1]
    return fwhm
```

```python
def print_fwhm_folder(filtname):
    fwhms = []
    for d in Path(f"../{objname}").glob(f"img_{filtname}_*"):

        fwhm = get_fwhm(d / "detection.cat")
        print(d, "\t", fwhm)

        fwhms.append(fwhm / u.arcsec)

    print("median\t\t", np.median(fwhms))
```

```python
def get_sky_brightness(foldername):
    bkg = CCDData.read(foldername / "bkg.fits", format="fits")
    
    mean, median, err = sigma_clipped_stats(bkg.data, stdfunc="mad_std")
    return median, err
```

```python
def print_fwhm_folder(filtname):
    fwhms = []
    bkgs = []
    bkg_errs = []
    print("folder\t\t\tfwhm\t\tback median\tbkg_err")
    for d in Path(f"../{objname}").glob(f"img_{filtname}_*"):

        fwhm = get_fwhm(d / "detection.cat")
        bkg_median, bkg_err = get_sky_brightness(d)
        print(f"{d}\t{fwhm:0.2f}\t{bkg_median:8.1f}\t{bkg_err:0.1f}")

        fwhms.append(fwhm / u.arcsec)
        bkgs.append(bkg_median)
        bkg_errs.append(bkg_err)

    fwhm = np.median(fwhms) * u.arcsec
    bkg_median = np.median(bkgs)
    bkg_err = np.median(bkg_err)
    print(f"median\t\t\t{fwhm:0.2f}\t{bkg_median:8.1f}\t{bkg_err:0.1f}")
```

# Show all images

```python
def sort_paths(paths):
    paths = [p for p in paths]
    idxs = [int(p.stem.split("_")[-1]) for p in paths]
    return np.array(paths)[np.argsort(idxs)]
```

```python
def show_cens(filtname):
    paths = sort_paths(Path(f"../{objname}").glob(f"img_{filtname}_*"))
    for d in paths:
        img = CCDData.read(d / "nobkg.fits")

        show_image(img[800:1200, 800:1200], log=True)
        fwhm = get_fwhm(d / "detection.cat")
        plt.title(str(d) + f" fwhm {fwhm:1.2f}")
```

```python
def show_all(filtname):
    paths = sort_paths(Path(f"../{objname}").glob(f"img_{filtname}_*"))
    for d in paths:
        img = CCDData.read(d / "nobkg.fits")

        show_image(img, log=True)
        fwhm = get_fwhm(d / "detection.cat")
        plt.title(str(d) + f" fwhm {fwhm:1.2f}")
```

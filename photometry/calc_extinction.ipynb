```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
atm_ext = pd.read_csv("../survey_data/gtc_extinction.csv", sep=" +")
```

```python
transmission_g = pd.read_csv("../survey_data/gtc_transmission_g.csv", sep=" +")
transmission_r = pd.read_csv("../survey_data/gtc_transmission_r.csv", sep=" +")
transmission_i = pd.read_csv("../survey_data/gtc_transmission_i.csv", sep=" +")
```

```python
atm_ext_interp = lambda x: np.interp(x, atm_ext.wavelength/10, atm_ext.extinction)
```

```python
def calc_ext(transmission):
    A = atm_ext_interp(transmission.wavelength)
    x = transmission.transmission  / transmission.wavelength**-1
    return np.sum(A * x ) / np.sum(x)
```

```python
calc_ext(transmission_g), calc_ext(transmission_r), calc_ext(transmission_i)
```

```python
ext = np.sum(atm_ext_interp(transmission_g.wavelength) * transmission_g.transmission) / np.sum(transmission_g.transmission)
```
